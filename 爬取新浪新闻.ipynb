## 爬取图片保存在本地

```python
# coding=utf-8
import urllib.request,re
import xlwt 
workbook = xlwt.Workbook(encoding = 'utf-8')
worksheet = workbook.add_sheet('图片链接',cell_overwrite_ok=True)
response = urllib.request.urlopen('http://www.imooc.com/course/list')
html = response.read().decode('utf-8')
# print(html)
listurl = re.findall(r'src=.+\.jpg', html)   # re库主要用在字符串匹配中
# 这里在查找src=...........jpg的图片
# print(listurl)
# for i in range(len(listurl)):
#     print(listurl[i][7:])
listurls = []
for each in listurl:
    listurls.append('http:' + each[5:])
#     print(listurls)
    i = 0
    for url in listurls:
        print(url)  # 输出图片的连接
        worksheet.write(i,0,listurls[i])
        response = urllib.request.urlopen(url)
        image = response.read()
        dir = 'E:\\ivy_machine_learning\\program\\莫烦学习\\爬虫图片\\' + str(i)+'.jpg' 
        f = open(dir, 'wb')	
        f.write(image)	
        i += 1

workbook.save(r'E:\ivy_machine_learning\program\莫烦学习\爬虫图片\图片路径.xls')
```
